# **Extracting Stock Data Using a Web Scraping**

Since APIs don't provide all the stock data we need, we'll have to use web scraping techniques with BeautifulSoup to extract it from a webpage

````python
!pip install pandas
!pip install requests
!pip install bs4
!pip install html5lib
!pip install lxml
!pip install plotly
````

````python
import pandas as pd
import requests
from bs4 import BeautifulSoup
````

````python
url='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html'

response=requests.get(url)
soup=BeautifulSoup(response.text,'html')
````

````python
table=soup.find_all('table')[0]
````

````python
raw_columns=table.find_all('th')
raw_columns
````

````python
column_names=[column.text.strip() for column in raw_columns]
````

````python
column_names
````

````python
df=pd.DataFrame(columns=column_names)
df
````

````python
column_data=table.find_all('tr')
````

````python
df.columns
````

````python
rows_list = []  # List to store all rows

# Loop through each row, process the data, and add it to the list
for row in column_data[:-1]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    # Only append rows with data, ignore empty lists
    if individual_row_data:
        rows_list.append(individual_row_data)

# Convert list of rows into a DataFrame at once using pd.concat
df = pd.concat([df, pd.DataFrame(rows_list, columns=df.columns)], ignore_index=True)
````

````python
df
````

````python
df[['Open', 'High', 'Low', 'Close*', 'Adj Close**', 'Volume']]=df[['Open', 'High', 'Low', 'Close*', 'Adj Close**', 'Volume']].map(lambda x: x.replace(',','')).astype(float)
````

````python
df['Date'] = pd.to_datetime(df['Date'], format='%b %d, %Y')
````
